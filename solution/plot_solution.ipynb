```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from tiles import domino
```

```python
phi = {}
for i in range(10):
    phi[i] = [i, i]
cnt = 10
for s in range(2):
    for i in range(10):
        for j in range(10):
            if j > i:
                if s == 0:
                    phi[cnt] = [i,j]
                else:
                    phi[cnt] = [j,i]
                cnt += 1
```

```python
sol = [line for line in open("sol.txt")]
imax = 20
jmax = 22
horz = []
vert = []
for l in sol:
    if "Th" in l:
        i = int(l.split("~")[1].split("_")[0].strip())
        j = int(l.split("~")[2].split("_")[0].strip())
        k = int(l.split("~")[3].split("_")[0].split()[0].strip())
        vert.append([i,j,phi[k][0],phi[k][1]])
    elif "Tv" in l:
        i = int(l.split("~")[1].split("_")[0].strip())
        j = int(l.split("~")[2].split("_")[0].strip())
        k = int(l.split("~")[3].split("_")[0].split()[0].strip())
        horz.append([i,j,phi[k][0],phi[k][1]])
```

```python
plt.figure(figsize=(10,10))
background = Image.new('RGBA', (20*390, 22*390), (255, 255, 255, 255))
for n in horz:
    d = domino(375, 750)
    d.add_dots(n[2],n[3])
    background.paste(d.image.rotate(90, expand=True), (390*(n[0]),390*n[1]))
for n in vert:
    d = domino(375, 750)
    d.add_dots(n[2],n[3])
    background.paste(d.image, (390*(n[0]),390*n[1]))
plt.imshow(background)
```
