```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = Image.open('IMG_20210324_112419.jpg')
im = im.crop((200,0,1800,2475))
im = ImageOps.grayscale(im)
im2 = im.resize((60,88))
im3 = im.resize((20,22))
a = 255-np.asarray(im2)
a
```

```python
bins = np.arange(11)
px_vals = bins-1
mn = a.min()
mx = a.max()
out = open("mary.dat","w")
imax = a.shape[0]
jmax = a.shape[1]
for i in range(imax):
    s = ""
    for j in range(jmax):
        px = 9*((a[i,j]-mn)/(mx-mn))
        s += "%d "%(px_vals[np.digitize(px,bins)])
    out.write(s[:-1]+"\n")
out.close()
```
